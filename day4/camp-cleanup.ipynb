## Part 1

```python
cleaningSchedule = 'input.txt'
```

```python
with open(cleaningSchedule) as f:
    lines = f.readlines()
```

```python
gptPairs = []
# Parse the input and store it in a list of tuples
gptPairs = []
for line in lines:
  range1, range2 = line.strip().split(",")
  start1, end1 = map(int, range1.split("-"))
  start2, end2 = map(int, range2.split("-"))
  gptPairs.append(((start1, end1), (start2, end2)))

# Count the number of pairs in which one range fully contains the other
count = 0
for (start1, end1), (start2, end2) in gptPairs:
  if start1 <= start2 and end1 >= end2:
    count += 1
  elif start2 <= start1 and end2 >= end1:
    count += 1

print(count)
```
